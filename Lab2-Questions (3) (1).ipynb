Create a 2-D array called myarray4 using arange() having 14 rows and 3 columns with start value = -1, step size 0.25 having. Split this array row wise into 3 equal parts and print the result. 

```python
# Write your answer here

import numpy as np

# Create a 2-D array myarray4 with 14 rows and 3 columns using arange
start_value = -1
step_size = 0.25
total_elements = 14 * 3  # 14 rows * 3 columns = 42 elements

# Creating the array
myarray4 = np.arange(start_value, start_value + step_size * total_elements, step_size).reshape(14, 3)

print("Original myarray4:\n", myarray4)

# Split the array row-wise into 3 equal parts
split_arrays = np.array_split(myarray4, 3)

# Print the result of the split
print("\nSplitted arrays:")
for idx, arr in enumerate(split_arrays):
    print(f"\nPart {idx + 1}:\n", arr)

```

#### Practice Exercise 2
Read the coordinates of the capital cities of the world from the country-capital-lat-long-population.csv file. Use NumPy to print the name and coordinates of the capital city closest to the Saudi Arabia capital - Riyadh.

Note that:

1. The Country Name for Saudi Arabia is given as Saudi Arabia in the data.
2. The ‘closeness’ of capital cities from the Saudi Arabia capital is based on the Euclidean distance of their coordinates to those of the Saudi Arabia capital.


Hints:

1. Use read_html() from the Pandas library to read the table.
2. Use the to_numpy() function of the Pandas DataFrame class to convert a DataFrame to a Numpy array
3. Use broadcasting to compute the euclidean distance of capital cities from Riyadh.

```python
import pandas as pd
import numpy as np

# Step 1: Read the CSV file
file_path = 'country-capital-lat-long-population.csv'  # Replace with your actual file path
data = pd.read_csv(file_path)

# Step 2: Check the column names and the first few rows
print("Columns in the DataFrame:", data.columns.tolist())
print(data.head())

# Step 3: Convert Latitude and Longitude to numeric, forcing errors to NaN
data['Latitude'] = pd.to_numeric(data['Latitude'], errors='coerce')
data['Longitude'] = pd.to_numeric(data['Longitude'], errors='coerce')

# Step 4: Extract coordinates of Riyadh
riyadh_row = data[data['Country'] == 'Saudi Arabia']  # Use the correct column name
riyadh_coords = riyadh_row[['Latitude', 'Longitude']].to_numpy().flatten()  # Get Riyadh's coordinates

# Check for NaN values in Riyadh's coordinates
print(f"Riyadh Coordinates: {riyadh_coords}")
if np.any(np.isnan(riyadh_coords)):
    print("Riyadh coordinates contain NaN values. Check the data.")
else:
    # Step 5: Extract coordinates of all capital cities
    capitals_coords = data[['Capital City', 'Latitude', 'Longitude']].to_numpy()
    print(f"Capitals Coordinates:\n{capitals_coords}")

    # Ensure all coordinates are valid and numeric
    print("Data Types:", capitals_coords.dtype)
    print("Riyadh Latitude:", type(riyadh_coords[0]), "Longitude:", type(riyadh_coords[1]))

    # Step 6: Calculate Euclidean distances
    distances = np.sqrt((capitals_coords[:, 1] - riyadh_coords[0]) ** 2 + 
                        (capitals_coords[:, 2] - riyadh_coords[1]) ** 2)
    print("Distances:", distances)

    # Step 7: Find the index of the closest capital city
    closest_index = np.argmin(distances)

    # Step 8: Print the name and coordinates of the closest capital city
    closest_capital = capitals_coords[closest_index]
    print(f"Closest capital city to Riyadh:")
    print(f"Name: {closest_capital[0]}")
    print(f"Coordinates: Latitude = {closest_capital[1]}, Longitude = {closest_capital[2]}")

```

#### Practical Exercise 2

Use NumPy to:

1. Print the names of the countries of the top 10 capital cities closest to the US capital - Washington DC.

2. Create and print a NumPy array containing the coordinates of the top 10 cities.

```python
import pandas as pd
import numpy as np

# Read the CSV file
data = pd.read_csv('/Users/reema/Downloads/country-capital-lat-long-population (1).csv')

# Convert Latitude and Longitude to numeric and drop rows with NaN values
data['Latitude'] = pd.to_numeric(data['Latitude'], errors='coerce')
data['Longitude'] = pd.to_numeric(data['Longitude'], errors='coerce')
data.dropna(subset=['Latitude', 'Longitude'], inplace=True)  # Remove rows with NaN coordinates

# Get Washington D.C.'s coordinates
dc_coords = data.loc[data['Capital City'] == 'Washington, D.C.', ['Latitude', 'Longitude']].values.flatten()

# Ensure we have valid coordinates for Washington, D.C.
if dc_coords.size == 0:
    raise ValueError("Washington, D.C. not found in the dataset.")

# Get coordinates of all capital cities
capitals_coords = data[['Country', 'Capital City', 'Latitude', 'Longitude']].values

# Calculate Euclidean distances to Washington D.C.
distances = np.sqrt((capitals_coords[:, 2] - dc_coords[0]) ** 2 + (capitals_coords[:, 3] - dc_coords[1]) ** 2)

# Get indices of the top 10 closest capital cities
closest_indices = np.argsort(distances)[:10]

# Print the names of the countries and their coordinates
print("Top 10 closest capital cities to Washington D.C.:")
for index in closest_indices:
    country = capitals_coords[index, 0]
    capital = capitals_coords[index, 1]
    coords = capitals_coords[index, 2:4]  # Get latitude and longitude
    print(f"{country} - {capital}: (Latitude: {coords[0]}, Longitude: {coords[1]})")

# Create and print a NumPy array containing the coordinates of the top 10 cities
top_10_coords = capitals_coords[closest_indices, 2:4]  # Extract only Latitude and Longitude
print("\nCoordinates of the top 10 closest capital cities:")
print(top_10_coords)

```
